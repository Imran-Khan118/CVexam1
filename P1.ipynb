```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

# Task 1


```python
image = cv2.imread('/content/33.jpg')

original_image = image[:, :image.shape[1] // 2]

# Section 1: Display the Original Image
plt.figure(figsize=(5, 5))
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")
plt.show()
```

```python
def apply_white_balance(img):
    result = cv2.cvtColor(img, cv2.COLOR_BGR2LAB)
    l, a, b = cv2.split(result)
    l = cv2.equalizeHist(l)
    result = cv2.merge((l, a, b))
    return cv2.cvtColor(result, cv2.COLOR_LAB2BGR)
```

```python

balanced_image = apply_white_balance(original_image)
gray_image = cv2.cvtColor(balanced_image, cv2.COLOR_BGR2GRAY)
equalized_image = cv2.equalizeHist(gray_image)
blurred_image = cv2.GaussianBlur(balanced_image, (5, 5), 0)
bilateral_filtered_image = cv2.bilateralFilter(balanced_image, 9, 75, 75)

lis = [original_image, balanced_image, equalized_image, blurred_image, bilateral_filtered_image]

# Plotting all images using subplots
plt.figure(figsize=(15, 10))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

# White Balanced Image
plt.subplot(2, 3, 2)
plt.imshow(cv2.cvtColor(balanced_image, cv2.COLOR_BGR2RGB))
plt.title("White Balanced Image")
plt.axis("off")

# Histogram Equalized Image
plt.subplot(2, 3, 3)
plt.imshow(equalized_image, cmap='gray')
plt.title("Histogram Equalized Image")
plt.axis("off")

# Gaussian Blurred Image
plt.subplot(2, 3, 4)
plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))
plt.title("Gaussian Blurred Image")
plt.axis("off")

# Bilateral Filtered Image
plt.subplot(2, 3, 5)
plt.imshow(cv2.cvtColor(bilateral_filtered_image, cv2.COLOR_BGR2RGB))
plt.title("Bilateral Filtered Image")
plt.axis("off")

# Show all images
plt.tight_layout()
plt.show()

#
```

# Task 2

```python


edges = []
for i in lis:

  edges.append(cv2.Canny(i, 50, 150))
# Display the Results
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(2, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

for x in range(len(lis)):
  plt.subplot(2, 3, x+2)
  plt.imshow(edges[x], cmap='gray')
  plt.title("Edges Detected")
  plt.axis("off")


plt.tight_layout()
plt.show()
```

I think edge detection from original image is working better than enhanced images

```python

gaussian_blurred = cv2.GaussianBlur(original_image, (5, 5), 0)
median_filtered = cv2.medianBlur(gaussian_blurred, 5)
bilateral_filtered = cv2.bilateralFilter(median_filtered, 9, 75, 75)

# Display the Results
plt.figure(figsize=(15, 5))

# Original Image
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))
plt.title("Original Image")
plt.axis("off")

# Median Filtered Image
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(median_filtered, cv2.COLOR_BGR2RGB))
plt.title("Median Filtered Image")
plt.axis("off")

# Bilateral Filtered Image
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(bilateral_filtered, cv2.COLOR_BGR2RGB))
plt.title("Bilateral Filtered Image")
plt.axis("off")

plt.tight_layout()
plt.show()


```
